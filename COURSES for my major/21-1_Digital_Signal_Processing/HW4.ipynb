Make a square signal at 1100 Hz and make a wave that samples it at 10000 frames per second.
If you plot the spectrum, you can see that most of the harmonics are aliased.
When you listen to the wave, can you hear the aliased harmonics? 

```python
import os

from thinkdsp import SquareSignal
from thinkdsp import decorate
```

```python
square = SquareSignal(1100).make_wave(duration=0.5, framerate=10000)
```

```python
square.make_spectrum().plot()
decorate(xlabel='Frequency (Hz)')
```

```python
square.make_audio()
```

fundamental frequency는 1100이고, harmonic은 3300, 5500, 7700...으로 예상된다.

하지만 5500이 아니라 4500, 7700이 아니라 2300, 9900이 아니라 100의 harmonic이 된다.

이것으로 충분한 sampling이 되지 않아 aliasing이 된 것을 알 수 있다.
