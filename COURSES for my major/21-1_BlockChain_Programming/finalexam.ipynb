````python
%%writefile src/CustomerAndOrder.sol
pragma solidity ^0.5.0;

contract Customer {
    struct Cus{
        uint id; //고객ID
        string name; //이름
        string ph; //전화번호
        string home; //집주소
    }
    mapping(address => Cus) cusInfo;
    
    address payable owner;
    constructor() public {
        owner = tx.origin;
    }

    function addCustomer(address addr,uint _id, string memory _name, string memory _ph, string memory _home) public {
         Cus memory customer=Cus(_id, _name, _ph, _home);
         cusInfo[addr]=customer;
    }
    
    function getHomeAddress(address addr) public view returns(string memory){
        Cus memory customer=cusInfo[addr];
        return customer.home;
    }
    function getId(address addr) public view returns (uint){
        Cus memory customer=cusInfo[addr];
        return customer.id;
    }  
}

contract Order {
    struct Ord{
        uint id; //주문ID
        string p; //상품명
        uint n; //개수
        uint pCost; //금액
        uint256 time; //시간
        string shipAddr; //배송지
    }
    mapping (address => uint) orderAmount; // 주문금액
    mapping (address => string) status; //상태
    mapping(address=>Ord) orderInfo; 
    mapping(uint=>address) addrByID;
    uint public orderCount=0;
    
    address payable owner;
    
    Customer customer;
    
    constructor() public {
        owner = msg.sender;
        customer = new Customer();
    }
    
    function placeOrder(address addr, uint _id, string memory _p, uint _n, uint _amount) payable public{
        require(msg.value == _amount);
        Ord memory order=Ord(_id,_p,_n,_amount,now,getHomeAddress(addr));
        orderInfo[addr]=order;
        status[addr]="Ordered";
        orderCount+=1;
        addrByID[_id]=addr;
        orderAmount[owner]+=_amount; 
    }
    
    function addCustomer(address addr, uint _id, string memory _name, string memory _ph, string memory _home) public {
        customer.addCustomer(addr, _id, _name, _ph, _home);
    }
    
    function getHomeAddress(address addr) public view returns (string memory){
        return customer.getHomeAddress(addr); 
    }
    
    function getStatus(address addr) public view returns (string memory){
         return status[addr];
    }
    
    function updateStatus(uint _id, string memory _s) public returns(string memory) {
         status[addrByID[_id]]=_s;
         return status[addrByID[_id]];
    }
    
    function getOrderItem(address addr) public view isOwner returns (uint, string memory, string memory, string memory){
        Ord memory order =orderInfo[addr];
        return (order.id,order.p,status[addr],order.shipAddr);
    }
    
    function getOrderByID(uint _id) public view isOwner returns (uint, string memory, string memory, string memory){
        Ord memory order =orderInfo[addrByID[_id]];
        return (order.id,order.p,status[addrByID[_id]],order.shipAddr);
    }
    
    function getNOrder() public view returns (uint){
        return orderCount;
    }
    
    function getTotalOrderAmount() public view returns (uint){
        return orderAmount[owner];
    }

    function queryBalance()view public isOwner returns(uint){
        return address(this).balance;
    }
    
    modifier isOwner {
        require(msg.sender == owner);
        _;
    }
}
````
