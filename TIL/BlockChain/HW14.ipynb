```python
%%writefile src/Rsp.sol
pragma solidity ^0.5.0;

contract Rsp{
    uint8 betA;
    uint8 betB;
    address addrA;
    address addrB;
    string matchResult;
    bytes32 winner;
    uint balance;
    uint betAmount; 
    
    mapping(address => uint256) balanceOf;
    mapping(string => address) playerAddr;
    
    address payable owner;
    constructor() public {
        owner = msg.sender;
    }
    
    function setAddr(address ad)public{
        addrA=address(this);
        addrB=ad;
    }
    
    function setA(uint amount) public payable{
        uint8 n=uint8((uint256(keccak256(abi.encodePacked(block.timestamp, block.difficulty)))%3));
        betA=n;
        betAmount=amount;
        playerAddr["betA"]=addrA;
        
    }
    
    function setB(uint8 n, uint amount)public payable{
        betB=n;
        betAmount=amount;
        playerAddr["betB"]=addrB;
    }
    
    function play() public {
        if(betA==betB) {
            matchResult="tie";
            winner="N";
        } else if(betA==1 && betB==2) {
            matchResult="B wins";
            winner="B";
        } else if(betA==1 && betB==3) {
            matchResult="A wins";
            winner="A";
        } else if(betA==2 && betB==1) {
            matchResult="A wins";
            winner="A";
        } else if(betA==2 && betB==3) {
            matchResult="B wins";
            winner="B";
        } else if(betA==3 && betB==1) {
            matchResult="B wins";
            winner="B";
        } else if(betA==3 && betB==2) {
            matchResult="A wins";
            winner="A";
        }
    }
    
    function distributeBetAmount() public {
        if (winner=="A") {
            balanceOf[playerAddr["betA"]]+=betAmount;
            balanceOf[playerAddr["betB"]]-=betAmount;
        } else if (winner=="B") {
            balanceOf[playerAddr["betB"]]+=betAmount;
            balanceOf[playerAddr["betA"]]-=betAmount;
        }
    }
    
    function getMatchResult() view public returns(string memory) {
        return matchResult;
    }
    function getBalanceOfThis() public view returns (uint){
        return address(this).balance;
    }
    function getBalanceOfbetA() public view returns (uint) {
        return balanceOf[addrA];
    }
     function getBalanceOfbetB() public view returns (uint) {
        return balanceOf[addrB];
    }
    
    function deposit(uint amount) payable public onlyOwner { 
        require(msg.value == amount);
        balanceOf[addrA]+=amount;
        balanceOf[addrB]+=amount;
    }

     modifier onlyOwner {
        require(msg.sender == owner);
        _;
    }
}
```
