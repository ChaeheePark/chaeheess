#### 5. autocorrelation을 이용하여 perfect_pitch에 포함되어 있는 음을 모두 순서대로 찾으세요.

autocorrelation에서 만든 estimate_fundamental을 이용하여 fundamental frequency를 찾는다.

```python
from autocorr import autocorr
```

```python
def estimate_fundamental(segment, low, high):
    lags, corrs = autocorr(segment)
    lag = np.array(corrs[low:high]).argmax() + low
    period = lag / segment.framerate
    frequency = 1 / period
    return frequency
```

estimate_fundamental 에서 correlation의 최댓값을 찾기위해서 lag의 low,high를 지정해주어야 해서 그래프를 하나씩 살펴보았다. 

1) 1.1 ~ 1.5 sec

```python
segment = wave.segment(start=1.1,duration=0.4)
lags, corrs = autocorr(segment)
plt.plot(lags, corrs)
plt.xlim([0, 500]) #확대
decorate(xlabel='Lag (index)', ylabel='Correlation')
```

```python
segment = wave.segment(start=1.1,duration=0.4)
freq = estimate_fundamental(segment,200,400)
freq
```

fundamental frequency는 130Hz이고 이 음계는 C3으로, 3옥타브 '도'음이다. 앞서 본 spectrum 분석과 같다.

<br><br>
2) 1.7 ~ 2.2 sec

```python
segment = wave.segment(start=1.7,duration=0.5)
lags, corrs = autocorr(segment)
plt.plot(lags, corrs)
plt.xlim([0, 500]) #확대
decorate(xlabel='Lag (index)', ylabel='Correlation')
```

#### 6. perfect_pitch를 분석해보고 디지털신호처리 관점에서 느낀점은? (코드 작성은 필요없는 문항임)


spectogram으로 몇개의 음계가 있는지 알아볼 수 있었다.

또한 그 음계가 어떤 주파수를 갖고있고 fundamental frequency로 음악상의 음계를 알아볼 수 있는 방법은
spectrum 분석방법과 autocorrelation의 방법이 있었는데,

spectrum 분석방법에서는 spectrum그래프에서 amplitude가 큰 peak 20개를 찍어본결과, 그 peak들의 주파수가 거의 harmonic하게 배수로 되어있는 것을 알아 볼 수 있었고, 이를 통해 음계를 확인하여 한 음을 구성하는 주파수들은 거의 조화롭게 이루어져 있고 듣기에 편안하다는 것을 알아낼 수 있었다.

autocorrelation 방법에서는 autocorrelation 그래프가 cos 의 그래프의 모양과 거의 비슷하게 correlation이 진행되는 것을 직접 확인할 수 있었다. 그래서 correlation이 가장 큰 lag로 fundamental frequency를 알아낼 수 있었다.

이처럼 디지털신호처리로 단순히 주파수의 모양, 크기 뿐만아니라 음계와 몇개의 음이 있는지 까지 알아내는 것을 할 수 있다. 이 방법을 적용하여 노래나 클래식 음악도 분석하면 흥미로울 것 같다고 느꼈고, 직접 만든 signal을 통해 주파수를 이용하면 짧은 음악도 만들수 있을것 같다고 느꼈다.
